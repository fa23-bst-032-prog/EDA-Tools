```python
from IPython.display import display, Markdown
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, chi2_contingency

# Set style
sns.set(style="whitegrid")

# Load dataset
df = pd.read_csv("ecommerce_dataset.csv")

# Heading function
def heading(text):
    display(Markdown(f"## **{text}**"))

# ============================
# 1. Basic Overview
# ============================
heading("First 5 Rows of the Dataset")
display(df.head())

heading("Dataset Information")
df.info()

heading("Summary Statistics")
display(df.describe())

heading("Missing Values")
display(df.isnull().sum())

# ============================
# 2. Sales Insights
# ============================
# Total Revenue = price * quantity * (1 - discount)
df["revenue"] = df["price"] * df["quantity"] * (1 - df["discount"])

heading("Top 5 Customers by Revenue")
top_customers = df.groupby("customer_id")["revenue"].sum().sort_values(ascending=False).head(5)
display(top_customers)

heading("Revenue by Category")
rev_by_cat = df.groupby("category")["revenue"].sum().sort_values(ascending=False)
display(rev_by_cat)

plt.figure(figsize=(8,5))
sns.barplot(x=rev_by_cat.index, y=rev_by_cat.values, palette="plasma")
plt.title("Revenue by Product Category", fontsize=14, weight="bold")
plt.ylabel("Revenue")
plt.xlabel("Category")
plt.show()

# ============================
# 3. Regional Insights
# ============================
heading("Revenue by Region")
rev_by_region = df.groupby("region")["revenue"].sum().sort_values(ascending=False)
display(rev_by_region)

plt.figure(figsize=(6,5))
sns.barplot(x=rev_by_region.index, y=rev_by_region.values, palette="coolwarm")
plt.title("Revenue by Region", fontsize=14, weight="bold")
plt.ylabel("Revenue")
plt.xlabel("Region")
plt.show()

# ============================
# 4. Payment Method Analysis
# ============================
heading("Revenue by Payment Method")
rev_by_payment = df.groupby("payment_method")["revenue"].sum().sort_values(ascending=False)
display(rev_by_payment)

plt.figure(figsize=(7,5))
sns.barplot(x=rev_by_payment.index, y=rev_by_payment.values, palette="Set2")
plt.title("Revenue by Payment Method", fontsize=14, weight="bold")
plt.ylabel("Revenue")
plt.xlabel("Payment Method")
plt.show()

# ============================
# 5. Hypothesis Testing
# ============================
heading("Hypothesis Testing")

# Example 1: Do Electronics and Sports categories have different average revenue per order?
electronics = df[df["category"]=="Electronics"]["revenue"]
sports = df[df["category"]=="Sports"]["revenue"]

t_stat, p_val = ttest_ind(electronics, sports, equal_var=False)

display(Markdown(f"**T-test between Electronics and Sports Revenue**"))
display(Markdown(f"T-statistic: {t_stat:.3f}, P-value: {p_val:.3f}"))

if p_val < 0.05:
    display(Markdown("✅ We reject the null hypothesis: Revenue distribution is significantly different between Electronics and Sports."))
else:
    display(Markdown("❌ We fail to reject the null hypothesis: No significant difference in revenues."))

# Example 2: Is region associated with payment method? (Chi-Square test)
contingency = pd.crosstab(df["region"], df["payment_method"])
chi2, p, dof, exp = chi2_contingency(contingency)

display(Markdown(f"**Chi-Square Test of Independence (Region vs Payment Method)**"))
display(Markdown(f"Chi2: {chi2:.3f}, P-value: {p:.3f}"))

if p < 0.05:
    display(Markdown("✅ We reject the null hypothesis: Region and Payment Method are dependent."))
else:
    display(Markdown("❌ We fail to reject the null hypothesis: No relationship between Region and Payment Method."))

# ============================
# 6. Time-based Analysis
# ============================
heading("Revenue Trend Over Time")
df["order_date"] = pd.to_datetime(df["order_date"])
rev_by_date = df.groupby(df["order_date"].dt.date)["revenue"].sum()

plt.figure(figsize=(10,5))
rev_by_date.plot(kind="line", marker="o", color="green")
plt.title("Revenue Trend Over Time", fontsize=14, weight="bold")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```
